```python
#Generate continuous and discrete systems in python
#unit impulse

import matplotlib.pyplot as plt 
import numpy as np
d=[]
n=np.arange(-10,10,1)
for i in n:
    if (i==0):
        d.append(1)
    else:
        d.append(0)
plt.plot(n,d)
```

```python
#unit step

import matplotlib.pyplot as plt
import numpy as np
d=[]
n=np.arange(-10,10,.1)
for i in n:
    if(i>=0):
        d.append(1)
    else:
        d.append(0)
plt.plot(n,d)            
```

```python
#exponential 
import matplotlib.pyplot as plt
import numpy as np
d=[]
n=np.arange(-10,10,0.1)
for i in n:
    d.append(np.exp(i))
plt.plot(n,d)    
```

```python
#sinusoidal
import matplotlib.pyplot as plt
import numpy as np
d=[]
n=np.arange(-10,10,0.1)
for i in n:
    d.append(np.sin(i))
plt.plot(n,d)
```

```python
#ramp
import matplotlib.pyplot as plt
import numpy as np
d=[]
n=np.arange(-10,10,0.1)
for i in n:
    d.append(2*i)
plt.plot(n,d)    
```

```python
#square
import matplotlib.pyplot as plt
import numpy as np
d=[]
n=np.arange(-10,10,0.1)
for i in n:
    if(i%2<=1):
        d.append(1)
    else:
        d.append(-1)
plt.plot(n,d)        
```

```python
#random 
import matplotlib.pyplot as plt
import numpy as np
import random
d=[]
n=np.arange(-10,10,1)
for i in n:
    d.append(np.random.randint(-100,100))
plt.plot(n,d)        
   
```
