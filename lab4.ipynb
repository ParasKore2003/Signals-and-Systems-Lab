```python
import matplotlib.pyplot as plt
import numpy as np
n1=np.arange(0,10,1)
c1=[]
for i in n1:
    if i==0:
        c1.append(1)
    else:
        c1.append(0)
#plt.stem(n1,c1)     

n2=np.arange(0,10,1)
c2=[]
c2.append(c1[0])

c2.append(c1[1]+0.5*c2[0])
for i in range(2,10):
    c2.append(c1[i] + (0.5)*c2[i-1] - (0.1)*c2[i-2])
plt.stem(n2,c2)    
```
