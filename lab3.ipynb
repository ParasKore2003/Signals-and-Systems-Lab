```python
# x[n]={1 0<=n<=5
#       0 otherwise}

import numpy as np
import matplotlib.pyplot as plt
n1=np.arange(-10,10,1)
c1=[]
for i in n1:
    if i<0:
        c1.append(0)
    elif i>5:
        c1.append(0)
    else:
        c1.append(1)
#plt.stem(n1,c1)
#plt.title('h[n]')
#plt.xlabel('n',fontsize=15)
#plt.ylabel('h[n]',fontsize=15)
#plt.grid()

n2=np.arange(-15,15,1)
c2=[]
for i in n2:
    if i<0:
        c2.append(0)
    elif i>5:
        c2.append(0)
    else:
        c2.append(1)
#plt.stem(n2,c2)
#plt.title('x[n]')
#plt.xlabel('n',fontsize=15)
#plt.ylabel('x[n]',fontsize=15)
#plt.grid()


n3=np.arange(-15,15,1)
c3=[]
for i in n3:
    z=0
    k=0
    for j in n3:
        z+=c2[j]*c2[i-j]  
    
    c3.append(z)
plt.stem(n3,c3)
plt.title('y[n]')
plt.xlabel('n',fontsize=15)
plt.ylabel('y[n]',fontsize=15)
plt.grid()
```
