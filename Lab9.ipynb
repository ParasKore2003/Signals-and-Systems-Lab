```python
import matplotlib.pyplot as plt
import numpy as np
import math as mt
f = 1
T = 1 / f 
t = np.arange(0, T, 0.01)
ts1 = np.arange(0, T, 0.1) 
b = np.sin(2 * np.pi * f * t) 
c = np.sin(2 * np.pi * f * ts1)
plt.plot(t,b,"g")
plt.stem(ts1,c,"k")
```

```python
c = np.sin(2 * np.pi * f * ts1)
plt.plot(ts1,c,"k")
```

```python
ts2 = np.arange(0, T, 0.05) 
d = np.sin(2 * np.pi * f * ts2)
plt.plot(ts2,d,"k")
```

```python
ts3 = np.arange(0, T, 0.08) 
e = np.sin(2 * np.pi * f * ts3)
plt.plot(ts3,e,"k")
```
