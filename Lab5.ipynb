```python
#Experiment 5

# Consider a periodic signal x(t) with period T as described below:
#      x(t) = 1   ,  |t| <= T1
#           = 0   ,  T1 < |t| <= T/2

#1. Choose a specific values of T in term of T1 and compute the fourier series coefficients and plot them.
#2. Also compute and plot the approximation signal x_M(t) by computing 2M+1 term approximation using fourier series.
#3. Compute the percentage of energy in error signal e_M(t) = x(t) - x_M(t)in one period for differnet values of M.
```

```python
#x[n]
import numpy as np
import matplotlib.pyplot as plt
X=[]
T1=20
T=2*T1
w=(2*np.pi)/T

for k in range(0,40,1):
    if(10<=k<=30):
        X.append(1)
    else:
        X.append(0)
#plt.plot(X)

#ck
ck=[]
for k in range(0,40,1):
    a=0
    for n in range(0,40,1):
        a=a+(X[n]*np.exp(complex(0,-k*n*w)))
    ck.append(a/T)
#plt.plot(ck)

X_M=[]
for k in range(0,40,1):
    b=0
    for n in range(0,40,1):
        b+=(ck[n]*np.exp(complex(0,-k*n*w)))
    X_M.append(b)
plt.plot(np.abs(X_M))    

#Error Term
E=[]
for k in range(0,40,1):
    E.append(X[k]-X_M[k])
error=0
for k in range(0,40,1):
    error+=E[k]
print(np.abs(error))
```
