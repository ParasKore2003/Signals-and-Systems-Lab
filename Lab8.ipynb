```python
#Consider a periodic signal X[n] with period N as described below:
#X[n]={ 0<=n<=N1-1
#       N1-1<=n<=N-1 }
#1) Find the response of this signal through an LTI system with
# a) y[n]={x[n] + x[n-1]}/2
# b) y[n]={x[n] - x[n-1]}/2

#2)Also plot the frequency response of the systems
```

```python
import numpy as np
import matplotlib.pyplot as plt

N1 = 20
N = 2 * N1
X = []
for n in range(0,N,1):
    if(0 <= n <= N1-1):
        X.append(1)
    else:
        X.append(0)
#plt.stem(X)
Y1 = []
for n in range(0,N,1):
    b = ( ( X[n] + X[n-1] ) / 2 )
    Y1.append(b)

#plt.stem(Y1)

Y2 = []
for n in range(0,N,1):
    c = ( ( X[n] - X[n-1] ) / 2 )
    Y2.append(c)
plt.stem(Y2)    
```

```python
from scipy.signal import freqz
b = [ 0.5, 0.5]
a = [1]
w,h = freqz( b, a )
plt.plot( w, abs(h))
```
