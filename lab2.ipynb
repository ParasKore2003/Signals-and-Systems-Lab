```python
#y(t)=x(t-5)
import matplotlib.pyplot as plt
import numpy as np
x=[]
y=[]
n=np.arange(-10,20,1)
for i in n:
    if(i>=0):
        x.append(i)
    else:
        x.append(0)
plt.plot(n,x)

plt.plot(n+5,x)
```

```python
#y(t)=x(2t)  & y(t)=x(0.2t)

import matplotlib.pyplot as plt
import numpy as np
x=[]
y=[]
n=np.arange(-10,20,1)
for i in n:
    if(i>=0):
        x.append(i)
    else:
        x.append(0)
plt.plot(n,x)
plt.plot(n/2,x)
plt.plot(n/.2,x)
```
